Simple Linear Regression

```python
import pandas as pd

df = pd.read_csv(".\data\Salary_Data_LR.csv")
df.head()
```

```python
from sklearn.linear_model import LinearRegression

X = df[['YearsExperience']]
Y = df['Salary']

model = LinearRegression()
model.fit(X, Y)
```

```python
predicted_salary = model.predict(pd.DataFrame([[4.2]], columns=['YearsExperience']))
print(f"Predicted salary for 4.2 years of experience: Rs.{predicted_salary[0]:.2f}/-")
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, Y, label='Data points')

plt.plot(X, model.predict(X), color='red', label='Regression line')

plt.xlabel('Years of Exp')
plt.ylabel('Salary (Rs)')
plt.legend()
plt.show()
```

```python
print(f"Slope of SLR model (m or Rate of change): {model.coef_[0]:.4f}")
print(f"Intercept of SLR model (c): {model.intercept_:.4f}")
```
